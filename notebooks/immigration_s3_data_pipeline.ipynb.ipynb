```python
import boto3
```

```python
s3 = boto3.client("s3")
s3.list_buckets()
```

```python
bucket_name = "immigration-data-kartheekp"
file_key = "Raw_Data/US Immigration Statistics (Ver 1.7.23).csv"
```

```python
response = s3.get_object(
    Bucket = bucket_name,
    Key = file_key
)
```

```python
from io import StringIO
import pandas as pd

csv_text = response["Body"].read().decode("utf-8")
df = pd.read_csv(StringIO(csv_text))
```

```python
df
```

Data Pre-processing:
1.Column inspection
2.Decide What Columns We Care About - remove unnecessary columns  - Here we are keeping as all are important ...
3.Rename Columns (Clarity & Professionalism)
4.Fix Data Types
5.Handle Missing Values

```python
#1.Column inspection
df.columns
```

```python
#3.Rename Columns (Clarity & Professionalism)
df = df.rename(columns = {
    "Year" : "year",
    "Immigrants Obtaining Lawful Permanent Resident Status" : "lpr",
    "Refugee Arrivals" : "refugee_arrivals",
    "Noncitizen Apprehensions": "noncitizen_apprehensions",
    "Noncitizen Removals" : "noncitizen_removals",
    "Noncitizen Returns" : "noncitizen_returns"
})
```

```python
#4.Fix Data Types
df.dtypes
```

```python
df['year'] = df['year'].astype(int)

numeric_cols = [
    'lpr',
    'refugee_arrivals',
    'noncitizen_apprehensions',
    'noncitizen_removals',
    'noncitizen_returns'
]

for col in numeric_cols:
    df[col] = df[col].str.replace(",", "", regex=False)

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col],errors='coerce')
```

```python
df.dtypes
```

```python
#5.Handle Missing Values 
#Count missing values (column-wise)
df.isnull().sum()
```

converting cleaned file to csv inorder to upload from python to s3 in aws..

```python
cleaned_file = "immigration_cleaned.csv"
df.to_csv(cleaned_file,index = False)
```

Upload to s3

```python
processed_file_key = "Processed_Data/immigration_cleaned.csv"

s3.upload_file(
    Filename = cleaned_file,
    Bucket = bucket_name,
    Key = processed_file_key
)
```

**Now Python → AWS S3 (processed data) → Power BI → Dashboard
       ***S3 → Public/Pre-signed URL → Power BI (Web connector) -Choose Access Type - Temporary Pre-Signed URL.

*Concept: What Is a Pre-Signed URL?


“A temporary key to read one file from S3.”

No AWS credentials inside Power BI

URL expires automatically

Power BI just sees it as a web file

```python
#Generating presigned URl
presigned_url = s3.generate_presigned_url(
    ClientMethod = "get_object",
    Params = {
        "Bucket" : bucket_name,
        "Key" : processed_file_key
    },
    ExpiresIn = 3600 #url expires in 1 hr
)

presigned_url
```
